```python
# ===============================
# 2. Compute client-level aggregated features from final_df
# -------------------------------

# 2.1. Percentage of Male Products (using the Universe column; assuming label 'Men')
gender_purchases = (final_df.groupby("ClientID")["Universe"]
                    .value_counts(normalize=True)
                    .unstack(fill_value=0)
                    .reset_index())
if "Men" not in gender_purchases.columns:
    gender_purchases["Men"] = 0
gender_purchases = gender_purchases[["ClientID", "Men"]].rename(columns={"Men": "PercentageMaleProducts"})

# 2.2. Total Amount Spent
total_spent = (final_df.groupby("ClientID")["SalesNetAmountEuro"]
               .sum()
               .reset_index()
               .rename(columns={"SalesNetAmountEuro": "TotalAmountSpent"}))

# 2.3. Total Quantity Bought (using Quantity_sold)
total_quantity = (final_df.groupby("ClientID")["Quantity_sold"]
                  .sum()
                  .reset_index()
                  .rename(columns={"Quantity_sold": "TotalQuantity"}))

# 2.4. Average Amount Per Transaction
avg_transaction = (final_df.groupby("ClientID")["SalesNetAmountEuro"]
                   .mean()
                   .reset_index()
                   .rename(columns={"SalesNetAmountEuro": "AverageAmountPerTransaction"}))

# 2.5. Days Since Last Transaction
last_transaction = (final_df.groupby("ClientID")["TransactionDate"]
                    .max()
                    .reset_index()
                    .rename(columns={"TransactionDate": "LastTransactionDate"}))
last_transaction["DaysSinceLastTransaction"] = (pd.Timestamp.now(tz='UTC') - last_transaction["LastTransactionDate"]).dt.days
last_transaction = last_transaction[["ClientID", "DaysSinceLastTransaction"]]

# 2.6. Most Bought Brand (based on sum of Quantity_sold per brand)
most_bought_brand = (final_df.groupby(["ClientID", "Brand"])["Quantity_sold"]
                     .sum()
                     .reset_index())
most_bought_brand = (most_bought_brand.sort_values(by=["ClientID", "Quantity_sold"], ascending=[True, False])
                     .drop_duplicates(subset=["ClientID"]))
most_bought_brand = most_bought_brand[["ClientID", "Brand"]].rename(columns={"Brand": "MostBoughtBrand"})

# 2.7. Purchase Frequency (average days between transactions per client)
purchase_dates = (final_df.groupby("ClientID")["TransactionDate"]
                  .apply(lambda dates: sorted(pd.to_datetime(dates)))
                  .reset_index())
purchase_dates["AverageFrequency"] = purchase_dates["TransactionDate"].apply(
    lambda dates: np.mean(np.diff(dates)) / np.timedelta64(1, 'D') if len(dates) > 1 else np.inf
)
purchase_dates = purchase_dates[["ClientID", "AverageFrequency"]]

# 2.8. Number of Unique Products Bought
unique_products = (final_df.groupby("ClientID")["ProductID"]
                   .nunique()
                   .reset_index()
                   .rename(columns={"ProductID": "UniqueProductsBought"}))

# 2.9. Amount Spent by Quarter
# First, create a Quarter column in final_df if not already present
final_df["Quarter"] = final_df["TransactionDate"].dt.quarter
amount_by_quarter = (final_df.groupby(["ClientID", "Quarter"])["SalesNetAmountEuro"]
                     .sum()
                     .unstack(fill_value=0))
amount_by_quarter.columns = [f"AmountSpent_Q{col}" for col in amount_by_quarter.columns]
amount_by_quarter = amount_by_quarter.reset_index()

# 2.10. Quantity Bought by Quarter
quantity_by_quarter = (final_df.groupby(["ClientID", "Quarter"])["Quantity_sold"]
                       .sum()
                       .unstack(fill_value=0))
quantity_by_quarter.columns = [f"Quantity_Q{col}" for col in quantity_by_quarter.columns]
quantity_by_quarter = quantity_by_quarter.reset_index()

# 2.11. Amount Spent Percentage per Category
amount_spent_per_category = (final_df.groupby(["ClientID", "Category"])["SalesNetAmountEuro"]
                             .sum()
                             .unstack(fill_value=0))
amount_spent_percentage = amount_spent_per_category.div(amount_spent_per_category.sum(axis=1), axis=0).fillna(0)
amount_spent_percentage.columns = [f"AmountSpentPct_{col}" for col in amount_spent_percentage.columns]
amount_spent_percentage = amount_spent_percentage.reset_index()
```

```python
# ===============================
# 3. Combine all aggregated features into one DataFrame keyed by ClientID
# -------------------------------
client_features = gender_purchases.merge(total_spent, on="ClientID", how="outer")
client_features = client_features.merge(total_quantity, on="ClientID", how="outer")
client_features = client_features.merge(avg_transaction, on="ClientID", how="outer")
client_features = client_features.merge(last_transaction, on="ClientID", how="outer")
client_features = client_features.merge(most_bought_brand, on="ClientID", how="outer")
client_features = client_features.merge(purchase_dates, on="ClientID", how="outer")
client_features = client_features.merge(unique_products, on="ClientID", how="outer")
client_features = client_features.merge(amount_by_quarter, on="ClientID", how="outer")
client_features = client_features.merge(quantity_by_quarter, on="ClientID", how="outer")
client_features = client_features.merge(amount_spent_percentage, on="ClientID", how="outer")

# Replace any remaining NaNs with sensible defaults
client_features.fillna({
    "PercentageMaleProducts": 0,
    "TotalAmountSpent": 0,
    "TotalQuantity": 0,
    "AverageAmountPerTransaction": 0,
    "DaysSinceLastTransaction": np.inf,
    "AverageFrequency": np.inf,
    "UniqueProductsBought": 0,
    "MostBoughtBrand": "Unknown"
}, inplace=True)
```

```python
# ===============================
# 4. Merge the client-level features into final_df (by ClientID)
# -------------------------------
final_df = final_df.merge(client_features, on="ClientID", how="left")

# Optional: Display a few rows to verify the new columns
display(final_df)
```
