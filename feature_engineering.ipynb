```python
################################
# 3) EXPANDING (PAST) FEATURES
################################
def compute_past_features(df):
    df = df.sort_values("TransactionDate").reset_index(drop=True)
    
    cumulative_spent = 0.0
    cumulative_quantity = 0
    men_items = 0
    
    brand_counts = defaultdict(int)
    unique_products = set()
    
    prev_date = None
    accum_days = 0
    transaction_count = 0
    
    # Prepare columns
    CumulativeSpent_col = []
    CumulativeQuantity_col = []
    PercentageMaleProducts_col = []
    DaysSinceLastTransaction_col = []
    MostBoughtBrandSoFar_col = []
    UniqueProductsSoFar_col = []
    AverageAmountPerTransactionSoFar_col = []
    AverageFrequencySoFar_col = []
    
    for _, row in df.iterrows():
        spent = row["SalesNetAmountEuro"]
        qty = row["Quantity_sold"]
        current_brand = row["Brand"]
        current_product = row["ProductID"]
        current_date = row["TransactionDate"]
        
        if prev_date is not None:
            days_diff = (current_date - prev_date).days
        else:
            days_diff = 0
        
        cumulative_spent += spent
        cumulative_quantity += qty
        if row["Universe"] == "Men":
            men_items += qty
        brand_counts[current_brand] += qty
        unique_products.add(current_product)
        
        transaction_count += 1
        accum_days += days_diff
        
        CumulativeSpent_col.append(cumulative_spent)
        CumulativeQuantity_col.append(cumulative_quantity)
        total_items_so_far = cumulative_quantity
        perc_male = men_items / total_items_so_far if total_items_so_far > 0 else 0
        PercentageMaleProducts_col.append(perc_male)
        
        best_brand = max(brand_counts, key=brand_counts.get)
        MostBoughtBrandSoFar_col.append(best_brand)
        
        UniqueProductsSoFar_col.append(len(unique_products))
        
        avg_amount = cumulative_spent / transaction_count if transaction_count > 0 else 0
        AverageAmountPerTransactionSoFar_col.append(avg_amount)
        
        avg_freq = accum_days / (transaction_count - 1) if transaction_count > 1 else 1
        AverageFrequencySoFar_col.append(avg_freq)
        
        DaysSinceLastTransaction_col.append(days_diff)
        prev_date = current_date
    
    df["DaysSinceLastTransaction"] = DaysSinceLastTransaction_col
    df["CumulativeSpent"] = CumulativeSpent_col
    df["CumulativeQuantity"] = CumulativeQuantity_col
    df["PercentageMaleProductsSoFar"] = PercentageMaleProducts_col
    df["MostBoughtBrandSoFar"] = MostBoughtBrandSoFar_col
    df["UniqueProductsSoFar"] = UniqueProductsSoFar_col
    df["AverageAmountPerTransactionSoFar"] = AverageAmountPerTransactionSoFar_col
    df["AverageFrequencySoFar"] = AverageFrequencySoFar_col
    
    return df

final_df = (final_df
    .groupby("ClientID", group_keys=False)
    .apply(compute_past_features)
    .sort_values("TransactionDate")
    .reset_index(drop=True)
)
```

```python

################################
# 4) PRODUCT-LEVEL AGGREGATES
################################
product_totals = (
    final_df.groupby("ProductID")
    .agg(
        TotalQuantitySold=("Quantity_sold", "sum"),
        TotalSales=("SalesNetAmountEuro", "sum")
    )
    .reset_index()
)
product_totals["AveragePrice"] = product_totals["TotalSales"] / product_totals["TotalQuantitySold"]
final_df = final_df.merge(product_totals[["ProductID","AveragePrice"]], on="ProductID", how="left")
```

```python

################################
# 5) ADD BASIC TIME FEATURES
################################
final_df["DayOfWeek"] = final_df["TransactionDate"].dt.dayofweek
final_df["Month"] = final_df["TransactionDate"].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

final_df["Season"] = final_df["Month"].apply(get_season)
```

```python
from tqdm import tqdm
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

def compute_rfm_windows_per_client_vectorized(df, windows=[30, 60, 90]):
    """
    Compute RFM windows for a single client's transactions in df using a vectorized approach for recency.
    """
    df = df.sort_values("TransactionDate").reset_index(drop=True)
    # Aggregate per day (floor TransactionDate)
    daily_data = df.groupby(df["TransactionDate"].dt.floor("D")).agg(
        TxCount=("TransactionDate", "count"),
        Spent=("SalesNetAmountEuro", "sum")
    ).sort_index()
    
    results = {}
    dates = daily_data.index.values  # numpy array of dates (dtype=datetime64[D])
    
    for w in windows:
        # Rolling sums for Frequency and Monetary
        roll_sum = daily_data.rolling(f"{w}D").sum().fillna(0)
        freq_col = f"Frequency_{w}"
        monet_col = f"Monetary_{w}"
        recency_col = f"Recency_{w}"
        
        roll_sum[freq_col] = roll_sum["TxCount"]
        roll_sum[monet_col] = roll_sum["Spent"]
        
        # Vectorized recency calculation:
        nonzero_dates = daily_data.index[daily_data["TxCount"] > 0].values
        idxs = np.searchsorted(nonzero_dates, dates, side='right') - 1
        recency = np.full(len(dates), w)
        valid = idxs >= 0
        candidates = nonzero_dates[idxs[valid]]
        window_start = dates[valid] - np.timedelta64(w-1, 'D')
        valid_candidate = candidates >= window_start
        recency_valid = (dates[valid] - candidates).astype('timedelta64[D]').astype(int)
        recency[valid] = np.where(valid_candidate, recency_valid, w)
        
        roll_sum[recency_col] = recency
        results[w] = roll_sum[[freq_col, monet_col, recency_col]]
    
    combined = daily_data.copy()
    for w in windows:
        combined = combined.join(results[w][[f"Frequency_{w}", f"Monetary_{w}", f"Recency_{w}"]])
    
    combined = combined.rename_axis("DatePivot").reset_index()
    df["DatePivot"] = df["TransactionDate"].dt.floor("D")
    df_merged = df.merge(combined, on="DatePivot", how="left")
    df_merged.drop(columns=["DatePivot", "TxCount", "Spent"], errors="ignore", inplace=True)
    df_merged = df_merged.sort_values("TransactionDate").reset_index(drop=True)
    return df_merged

def compute_rfm_windows(final_df, windows=[30, 60, 90], n_jobs=-1):
    """
    Process RFM windows for all clients in parallel.
    """
    # Group data by ClientID
    grouped = list(final_df.groupby("ClientID"))
    
    # Use joblib.Parallel to process clients in parallel.
    results = Parallel(n_jobs=n_jobs, backend="loky", verbose=1)(
        delayed(compute_rfm_windows_per_client_vectorized)(group, windows=windows)
        for _, group in grouped
    )
    
    combined_df = pd.concat(results).sort_values("TransactionDate").reset_index(drop=True)
    return combined_df

# Example usage:
final_df = compute_rfm_windows(final_df, windows=[30, 60, 90], n_jobs=-1)
```

```python
from tqdm.notebook import tqdm
import pandas as pd
import numpy as np

def compute_rolling_category_pct_90d_per_client(df):
    """
    Calculate the rolling percentage of sales per category for a single client
    over a 90-day window.
    
    1. Convert TransactionDate to tz-naive and calculate the "floor" date (Date).
    2. Create a daily pivot (Date x Category) with the sum of SalesNetAmountEuro.
    3. Force the index to daily frequency (fill_value=0) and apply a 90-day rolling window.
    4. Calculate the fraction for each category (dividing by the total row sum).
    5. Rename the columns by adding the prefix "Rolling90Pct_".
    6. Merge with the original DataFrame (based on Date).
    """
    df = df.copy()
    # Ensure TransactionDate is tz-naive
    df["TransactionDate"] = df["TransactionDate"].dt.tz_localize(None)
    df = df.sort_values("TransactionDate").reset_index(drop=True)
    # Calculate the floor date
    df["Date"] = df["TransactionDate"].dt.floor("D")
    
    # Create the pivot table: index = Date, columns = Category, values = SalesNetAmountEuro
    daily_cat = df.groupby(["Date", "Category"])["SalesNetAmountEuro"].sum().unstack(fill_value=0)
    daily_cat.index = pd.to_datetime(daily_cat.index)
    # Force the index to daily frequency, filling with 0
    daily_cat = daily_cat.asfreq("D", fill_value=0).sort_index()
    
    # Apply the 90-day rolling window
    rolling_cat_90 = daily_cat.rolling("90D").sum().fillna(0)
    
    # Calculate the fraction for each category for each day
    total_spent_90 = rolling_cat_90.sum(axis=1)
    cat_pct_90 = rolling_cat_90.div(total_spent_90, axis=0).fillna(0)
    
    # Prepare for the merge: reset the index and rename the index to "DatePivot"
    cat_pct_90 = cat_pct_90.reset_index().rename(columns={"Date": "DatePivot"})
    # Rename the columns: "Rolling90Pct_{Category}"
    cat_pct_90 = cat_pct_90.rename(columns=lambda c: f"Rolling90Pct_{c}" if c != "DatePivot" else c)
    
    # Merge the result with the original DataFrame
    df["DatePivot"] = df["Date"]
    df_merged = pd.merge(df, cat_pct_90, on="DatePivot", how="left")
    df_merged.drop(columns=["Date", "DatePivot"], inplace=True)
    df_merged = df_merged.sort_values("TransactionDate").reset_index(drop=True)
    
    return df_merged

def compute_rolling_category_percentage_90d(final_df):
    """
    Apply the compute_rolling_category_pct_90d_per_client function for each client,
    showing a global progress bar.
    """
    # Group by ClientID
    groups = [group for _, group in final_df.groupby("ClientID", group_keys=False)]
    results = []
    for group in tqdm(groups, desc="Processing clients for 90D Rolling Category"):
        result = compute_rolling_category_pct_90d_per_client(group)
        results.append(result)
    final_df_out = pd.concat(results).sort_values("TransactionDate").reset_index(drop=True)
    return final_df_out

# Apply the function to the final DataFrame
final_df = compute_rolling_category_percentage_90d(final_df)
```
